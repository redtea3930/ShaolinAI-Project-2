```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
# Read the whole text.
text = open('../Resources/shaolin.txt
            ', encoding="utf8").read()
```

```python
# read the mask / color image taken from
coloring = np.array(Image.open('../Resources/shaolin.png'))
stopwords = set(STOPWORDS)
custom_stop_words = ['Shahar', 's', 'see', 'will', 'two', 'p']
[stopwords.add(n) for n in custom_stop_words]
```

```python
wc = WordCloud(background_color="white", max_words=5000, mask=coloring,
               stopwords=stopwords, max_font_size=150, random_state=42)
# generate word cloud
wc.generate(text)
```

```python
# create coloring from image
image_colors = ImageColorGenerator(coloring)
```

```python
# we could also give color_func=image_colors directly in the constructor
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.figure()
```

```python
wc.to_file('shaolin_wc.png')
```
